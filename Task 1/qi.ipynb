Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd 
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

Activity Labels

```python
activity = ["LAYING" , "SITTING","STANDING","WALKING", "WALKING_DOWNSTAIRS","WALKING_UPSTAIRS"]
dir_path = os.path.join("..","HAR", "Combined", "train")
t = 10
freq = 50
```

Loading Data

```python
activity_sample = []
for action in activity:
    path = os.path.join(dir_path, action)
    files = os.listdir(path)
    sel_file = os.path.join(path, files[-1])

    df = pd.read_csv(sel_file)

    df_first_500 = df[:t*freq]

    activity_sample.append(df_first_500.values)

activity_sample = np.array(activity_sample)
```

Plotting Data

```python
plt.figure(figsize=(18,6))
for i, sample in enumerate(activity_sample):
    plt.subplot(2, 3, i+1)
    plt.plot(sample)
    plt.title(activity[i])

plt.tight_layout()
plt.show()
```

#### The waveforms show some visible differences. The plots show that sitting has a distinct pattern and can be identified more easily. Laying and standing have almost flat signals, making them identifiable as a group. Walking, walking upstairs and downstairs have repeating patterns but look alike to each other, so they can be identified as a group.
